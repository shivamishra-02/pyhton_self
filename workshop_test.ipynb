```python
def convert_to_uppercase(strings):
    return [string.upper() for string in strings]


lower_case = ['shivam', 'mishra', 'ds-b']
upper_case = convert_to_uppercase(lower_case)
print(upper_case)
```

```python
from tkinter import*
root=Tk()
root.geometry("350x350")
root.title("calculator")
root.resizable(0,0)
input_text=StringVar()
input_frame=Frame(root,width=312,height=50,highlightthickness=1,bg="lightgrey")
input_frame.pack(side=TOP)
infield=Entry(input_frame,font=('arial',18,"bold"),textvariable=input_text,width=50,bg="lightblue",bd=5,justify=RIGHT)
infield.grid(row=0,column=0)
infield.pack(ipady=10)


btns_frame=Frame(root,width=312,height=272.5,bg="grey")
btns_frame.pack()

def btn_click(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)
    
def btn_equal():
    global expression
    result = str(eval(expression))
    input_text.set(result)
    expression=""
    
def btn_clear():
    global expression
    expression = ""
    input_text.set("")


#the first row will compormise of the buttons 'c'
clear = Button(btns_frame, text= 'C',fg= 'black',bg='grey',width=32,height=3,cursor='hand2' , command=lambda:btn_clear())
clear.grid(row = 0,column=0 ,columnspan = 3 ,padx=1 ,pady=1)

divide = Button(btns_frame, text ='/',
fg= 'black',width = 10,
height =3 ,bd =0 , bg='lightgrey', cursor = 'hand2',
command = lambda :btn_click('/')). grid(row = 0 ,column = 3, padx = 1,pady = 1)

seven =Button(btns_frame,text='7',fg='black',width='10',height='3',command = lambda:btn_click(7))
seven.grid(row=1,column=0,padx=1,pady=1)

eight =Button(btns_frame,text='8',fg='black',width='10',height='3',command = lambda:btn_click(8))
eight.grid(row=1,column=1,padx=1,pady=1)

nine =Button(btns_frame,text='9',fg='black',width='10',height='3',command = lambda:btn_click(9))
nine.grid(row=1,column=2,padx=1,pady=1)

four =Button(btns_frame,text='4',fg='black',width='10',height='3',command = lambda:btn_click(4))
four.grid(row=2,column=0,padx=1,pady=1)

five =Button(btns_frame,text='5',fg='black',width='10',height='3',command = lambda:btn_click(5))
five.grid(row=2,column=1,padx=1,pady=1)

six =Button(btns_frame,text='6',fg='black',width='10',height='3',command = lambda:btn_click(6))
six.grid(row=2,column=2,padx=1,pady=1)

one =Button(btns_frame,text='1',fg='black',width='10',height='3',command = lambda:btn_click(1))
one.grid(row=3,column=0,padx=1,pady=1)

two =Button(btns_frame,text='2',fg='black',width='10',height='3',command = lambda:btn_click(2))
two.grid(row=3,column=1,padx=1,pady=1)

three =Button(btns_frame,text='3',fg='black',width='10',height='3',command = lambda:btn_click(3))
three.grid(row=3,column=2,padx=1,pady=1)

multiply =Button(btns_frame,text='',fg='black',width='10',height='3',command = lambda:btn_click(""))
multiply.grid(row=1,column=3,padx=1,pady=1)

minus =Button(btns_frame,text='-',fg='black',width='10',height='3',command = lambda:btn_click("-"))
minus.grid(row=2,column=3,padx=1,pady=1)

plus =Button(btns_frame,text='+',fg='black',width='10',height='3',command = lambda:btn_click("+"))
plus.grid(row=3,column=3,padx=1,pady=1)
                            
                            
zero =Button(btns_frame,text='0',fg='black',width='22',height='3',command = lambda:btn_click(0))
zero.grid(row=4,column=0,padx=1,pady=1,columnspan = 2)
                            
point =Button(btns_frame,text='.',fg='black',width='10',height='3',command = lambda:btn_click("."))
point.grid(row=4,column=2,padx=1,pady=1)                            

equal =Button(btns_frame,text='=',fg='black',width='10',height='3', command=lambda:btn_equal())
equal.grid(row=4,column=3,padx=1,pady=1) 





root.mainloop()
```
