```python
exp=[1,0,2,4,6,8,10]
sal=[20000,0,25000,50000,75000,100000,140000]
plt.plot(exp,sal)
plt.ylabel("Salary")
plt.xlabel("Exp")
plt.title("Salary-exp Graph")
plt.grid()
plt.show()
```

```python
mistakes=[1,0,2,4,6,8,10]
sal=[40000,100000,35000,30000,25000,20000,15000]
plt.plot(mistakes,sal)
plt.ylabel("Salary")
plt.xlabel("Mistakes")
plt.title("Salary-mistakes Graph")
plt.grid()
plt.show()
```

```python
numbers=[1,2,3,4,5,6,7]
squares=[1,4,9,16,25,36,49]
plt.plot(numbers,squares)
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.title("Square-numbers graph")
plt.grid()
plt.show()
```

```python
numbers=[1,2,3,4,5,6,7]
cubes=[1,8,27,64,125,216,343]
sqaures=[1,4,9,16,25,36,49]
plt.plot(numbers,cubes,squares)
plt.ylabel("Cubes")
plt.xlabel("Numbers")
plt.title("Square-numbers graph")
plt.plot(numbers, marker = '<')
plt.plot(cubes, marker = '*',ms = 15, mec = 'hotpink', mfc = 'hotpink')
plt.plot(squares,marker = 'o', ms = 10, mec = 'r',mfc='skyblue')
plt.grid()
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
species_counts = iris['species'].value_counts()
sns.barplot(x=species_counts.index, y=species_counts.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset('iris')

# Create a correlation matrix of the numerical columns
corr_matrix = tips.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, )

# Display the plot
import matplotlib.pyplot as plt
plt.show()
```

```python
sns.lineplot(x='year',y='passengers',hue='month',data=a)
plt.show()
```

```python
sns.pairplot(a)
plt.show()
```

```python
corr = a.corr()
ax1=sns.heatmap(corr,annot=True)
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.barplot(x='year',y='passengers',data=a)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.set(style='darkgrid')
sns.lineplot(x='month',y='passengers',data=a,palette='solid')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='year',y='passengers',data=a,hue='month')
plt.grid()
plt.show()
```
